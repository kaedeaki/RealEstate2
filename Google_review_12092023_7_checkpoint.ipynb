# Json file to Pandas Data Freme

```python
import csv
import json


#read json file
with open('google_maps_results_by_company4000.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# wite to CSV
with open('google_maps_results_by_company4000_2.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)

    # write header
    header_written = False
    for company_name, records in data.items():
        for record in records:
            for item in record:
                if not header_written:
                    csv_writer.writerow(item.keys())
                    header_written = True

                # write record
                csv_writer.writerow(item.values())
```

# Collecting reviews using a list from CSV

```python
import csv
import os
from outscraper import ApiClient

def getreview(api_key, place_id, reviews_limit=3):
    api_client = ApiClient(api_key=api_key)
    results = api_client.google_maps_reviews(place_id, reviews_limit=reviews_limit)

    reviews_list = []
    for result in results:
        place_id = result['place_id']
        name = result.get('name','') # if 'name' is not existing, then return blank str

        for review_data in result.get('reviews_data', [])[:reviews_limit]:
            review_text = review_data.get('review_text', '')
            author_name = review_data.get('author_title', '')
            review_rating = review_data.get('review_rating', 0)

            review_info = {
                'Company Name': name,
                'Place ID': place_id,
                'Author': author_name,
                'Rating': review_rating,
                'Review Text': review_text
            }

            reviews_list.append(review_info)

    return reviews_list


def collect_reviews_from_csv(api_key, csv_file_path, reviews_limit_column=2, checkpoint_file=None):
    # with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
    #     csv_reader = csv.reader(csvfile)
    #     next(csv_reader)

    all_reviews = []

    # If a checkpoint file is specified, resume where you left off
    if checkpoint_file is not None and os.path.exists(checkpoint_file):
        with open(checkpoint_file, 'r') as checkpoint:
            last_processed_line = int(checkpoint.readline())
    else:
        last_processed_line = 0

    with open(csv_file_path, 'r', encoding='utf-8') as csvfile:
        csv_reader = csv.reader(csvfile)
        next(csv_reader)

        line_number = 0
        for row in csv_reader:
            if line_number < last_processed_line:
                line_number += 1
                continue

            company_name, place_id, reviews_limit_str = row[0], row[1], row[reviews_limit_column]
            reviews_limit = int(float(reviews_limit_str)) if reviews_limit_str else 0
            print(f"\nCollecting reviews for {company_name} (Place ID: {place_id}) reviews:{reviews_limit}")
            reviews = getreview(api_key, place_id, reviews_limit)
            all_reviews.extend(reviews)

            # update the checkpoint
            with open(checkpoint_file, 'w') as checkpoint:
                checkpoint.write(str(line_number + 1))

            line_number += 1

    #save to CSV
    csv_output_file_path = 'text_based_reviews_4000builder_4.csv'
    with open(csv_output_file_path, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Company Name', 'Place ID', 'Author', 'Rating', 'Review Text']
        csv_writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        csv_writer.writeheader()
        csv_writer.writerows(all_reviews)

# Specify checkpoint file path
checkpoint_file_path = 'checkpoint.txt'

api_key = 'REDACTED'
csv_file_path = '/content/drive/MyDrive/NLP_exercise_12252023/Sharpest_minds_Khaled/RealEstate2_all/output_dataset/output_dataset_batch_4_clean.csv'
collect_reviews_from_csv(api_key, csv_file_path, checkpoint_file=checkpoint_file_path)
```
