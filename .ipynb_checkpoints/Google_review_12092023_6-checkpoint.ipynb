```python
import csv
import json
from outscraper import ApiClient

def get_google_maps_results(api_key, csv_filename):
    api_client = ApiClient(api_key=api_key)

    try:
        with open(csv_filename, 'r', encoding='utf-8') as csv_file:
            reader = csv.DictReader(csv_file)
            place_id_list = [row['place_id'] for row in reader]

        all_results = {}

        for place_id in place_id_list:
            try:
                result = api_client.google_maps_search(place_id, limit=3)
                #all_results[company_name] = result
                for result in results:
                    place_id = result['place_id']
                    name = result['company_name']
                    
                    print(f"\nReviews for {name} (Place ID: {place_id}):\n")
    
                    for review_data in result.get('reviews_data', []):
                        review_text = review_data.get('review_text', '')
                        author_name = review_data.get('author_title', '')
                        review_rating = review_data.get('review_rating', 0)
        
                        print(f"Author: {author_name}")
                        print(f"Rating: {review_rating}")
                        print(f"Review Text: {review_text}\n")
                    
            except Exception as e:
                print(f"Error for {place_id}: {str(e)}")

        return all_results
    except Exception as e:
        print(f"Error: {str(e)}")
        return None

def create_csv_and_json(api_key, input_csv_filename, output_csv_filename, output_json_filename):
    results = get_google_maps_results(api_key, input_csv_filename)

    if results:
        with open(output_csv_filename, 'w', encoding='utf-8', newline='') as csv_file:
            fieldnames = ['company_name', 'result']  
            writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
            writer.writeheader()

            for company_name, result in results.items():
                writer.writerow({'company_name': company_name, 'result': result})
        
        with open(output_json_filename, 'w', encoding='utf-8') as json_file:
            json.dump(results, json_file, ensure_ascii=False, indent=2)

        print(f"CSV file '{output_csv_filename}' and JSON file '{output_json_filename}' created successfully.")
    else:
        print("CSV and JSON creation failed.")

api_key = 'YTVkNzRhOWRhMmFlNDFkYzkwOWFhZWY2ZTYxNzQ1NTJ8OGNhYjBjZWNmNw'
input_csv_filename = 'output_results.csv'
output_csv_filename = 'output_results2.csv'
output_json_filename = 'output_results2.json'

create_csv_and_json(api_key, input_csv_filename, output_csv_filename, output_json_filename)
```

```python
import json
from outscraper import ApiClient

def scrape_reviews(api_key, json_filename, output_json_filename):
    api_client = ApiClient(api_key=api_key)

    try:
        with open(json_filename, 'r', encoding='utf-8') as json_file:
            data_list = json.load(json_file)

        all_reviews = {}

        for result in data_list:
            company_name = result.get('company_name', '')
            place_id = result.get('result', {}).get('place_id', '')

            if company_name and place_id:
                try:
                    reviews = api_client.google_maps_reviews(place_id, reviews_limit=3)
                    all_reviews[company_name] = {'place_id': place_id, 'reviews': reviews}
                except Exception as e:
                    print(f"Error for {company_name}: {str(e)}")

        with open(output_json_filename, 'w', encoding='utf-8') as output_json_file:
            json.dump(all_reviews, output_json_file, ensure_ascii=False, indent=2)

        print(f"Reviews scraped and saved to '{output_json_filename}' successfully.")
    except Exception as e:
        print(f"Error: {str(e)}")

api_key = 'YTVkNzRhOWRhMmFlNDFkYzkwOWFhZWY2ZTYxNzQ1NTJ8OGNhYjBjZWNmNw'
json_filename = 'output_results.json'
output_json_filename = 'reviews_output.json'

scrape_reviews(api_key, json_filename, output_json_filename)
```
