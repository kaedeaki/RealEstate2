# NER

```python
# import flair library for NER
from flair.data import Sentence
from flair.nn import Classifier
from flair.models import SequenceTagger
 #Load the model
tagger = SequenceTagger.load('ner')
sentence = Sentence('George Washington went to Washington.')
tagger.predict(sentence)

print(sentence.to_tagged_string())

for entity in sentence.get_spans('ner'):
    print(entity)
```

```python
#function for NER
def apply_ner(text):
    sentence = Sentence(text)
    tagger.predict(sentence)
    entities = sentence.to_dict(tag_type='ner')['entities']
    return entities
```

```python
# change the data type to string in Review Text column
df['Review_Text'] = str(df['Review Text'])
```

```python
#apply NER to entire dataset
df['entities'] = df['Review Text'].apply(apply_ner)

#save the result
df.to_csv('/content/drive/MyDrive/NLP_exercise_12252023/Sharpest_minds_Khaled/RealEstate2_all/text_based_reviews/merge_file/ner_result.csv', index=False)
```

```python
# import pandas and ast
import pandas as pd
import ast  # convert to list

df = pd.read_csv('/content/drive/MyDrive/NLP_exercise_12252023/Sharpest_minds_Khaled/RealEstate2_all/text_based_reviews/merge_file/ner_result.csv')
```

```python
# function to clean the reviews (eliminate unnecessary words and change the sentence.)
def clean_review_text(text):
    text = text.replace('\n', ' ').replace('\r', '')
    return text.strip()

# clean the sentences of column 'Review Text'. Apply the function of "clean_review_text".
df['cleaned_review'] = df['Review Text'].apply(clean_review_text)

# check the data frame
df.head()
```

```python
#convert str to list in the column of entities
df['entities'] = df['entities'].apply(lambda x: ast.literal_eval(x))
```

```python
# function for extracting builder name
def extract_buildername(entities):
    for entity in entities:
        if 'labels' in entity and entity['labels'][0]['value'] == 'PER':
            return entity['text']
        else:
            return None
```

```python
# function for extracting builder company
def extract_company(entities):
    companies = [entity['text'] for entity in entities if 'labels' in entity and entity['labels'][0]['value'] == 'ORG']
    return companies if companies else None
```

```python
# function for extracting builder location
def extract_location(entities):
    locations = [entity['text'] for entity in entities if 'labels' in entity and entity['labels'][0]['value'] == 'LOC']
    return locations if locations else None
```

```python
# function for extracting builder skills and characteristics
def extract_skills(entities, review_text):
    # extract the keyword of the specific characteristics
    skills = []
    for entity in entities:
        if 'labels' in entity and entity['labels'][0]['value'] == 'MISC':
            skills.append(entity['text'])

    # extract the specific keyword from review_text
    if "three-story" in review_text.lower():
        skills.append("three-story building")
    if "two-story" in review_text.lower():
        skills.append("two-story building")
    if "one-story" in review_text.lower():
        skills.append("one-story building")
    if "garage" in review_text.lower():
        skills.append("garage construction")
    if "basement" in review_text.lower():
        skills.append("basement construction")
    if "pool" in review_text.lower():
        skills.append("pool construction")
    if "garden" in review_text.lower():
        skills.append("garden construction")
    if "roof" in review_text.lower():
        skills.append("roof construction")
    if "elevator" in review_text.lower():
        skills.append("elevator construction")
    if "fireplace" in review_text.lower():
        skills.append("fireplace construction")
    if "security" in review_text.lower():
        skills.append("security construction")
    if "laundry" in review_text.lower():
        skills.append("laundry construction")
    if "storage" in review_text.lower():
        skills.append("storage construction")
    if "heating" in review_text.lower():
        skills.append("heating construction")
    if "cooling" in review_text.lower():
        skills.append("cooling construction")
    if "air conditioning" in review_text.lower():
        skills.append("air conditioning construction")
    if "water heater" in review_text.lower():
        skills.append("water heater construction")
    if "solar panel" in review_text.lower():
        skills.append("solar panel construction")
    if "panoramic view" in review_text.lower():
        skills.append("panoramic view construction")
    if "outdoor space" in review_text.lower():
        skills.append("outdoor space construction")
    if "balcony" in review_text.lower():
        skills.append("balcony construction")
    if "designable" in review_text.lower():
        skills.append("designable construction")
    if "interior design" in review_text.lower():
        skills.append("interior design construction")
    if "exterior design" in review_text.lower():
        skills.append("exterior design construction")
    if "renovation" in review_text.lower():
        skills.append("renovation construction")
    if "simple" in review_text.lower():
        skills.append("simple construction")
    if "sustainable" in review_text.lower():
        skills.append("sustainable construction")
    if "durable" in review_text.lower():
        skills.append("durable construction")

    return skills if skills else None
```

```python
# add each columns with new data
df['builder_names'] = df['entities'].apply(extract_buildername)
df['companies'] = df['entities'].apply(extract_company)
df['locations'] = df['entities'].apply(extract_location)
df['skills'] = df.apply(lambda row: extract_skills(row['entities'], row['cleaned_review']), axis=1)

# show the result
print(df[['cleaned_review', 'builder_names', 'companies', 'locations', 'skills']].head(30))
```

```python
# filtering the builders which specialize "garage construction"
garage_specialists = df[df['skills'].apply(lambda x: 'garage construction' in (x or []))]

# show the result
print(garage_specialists[['Company Name', 'builder_names', 'skills']])
```

#Flair version of sentiment analysis  

```python
from flair.models import TextClassifier
from flair.data import Sentence

# load sentiment analysis model from Flair
classifier = TextClassifier.load('sentiment')

# function to apply sentiment analysis
def get_sentiment(text):
    sentence = Sentence(text)
    classifier.predict(sentence)
    sentiment_label = sentence.labels[0].value  # label with positive/negative
    sentiment_score = sentence.labels[0].score  # confident level(range 0~1）
    return sentiment_label, sentiment_score

# calculate the sentiment score and add to the dataframe
df['sentiment'], df['sentiment_score'] = zip(*df['cleaned_review'].apply(get_sentiment))

# show the result
df[['cleaned_review', 'sentiment', 'sentiment_score']].head()
```

```python
# save the data frame as CSV
df.to_csv('/content/drive/MyDrive/NLP_exercise_12252023/Sharpest_minds_Khaled/RealEstate2_all/text_based_reviews/merge_file/flair_review_with_sentiment.csv', index=False)

# show the result of sentiment analysis
df[['cleaned_review', 'sentiment', 'sentiment_score']].head()
```

```python
# decide the threshhold. Let's say more than 0.7 is positive review.
threshold = 0.7

# extract builers which have only positive sentiment scores
positive_reviews = df[df['sentiment_score'] >= threshold]

# filtering with skill(positive review with the companies specialized garage construction for example)
positive_garage_specialists = positive_reviews[positive_reviews['skills'].apply(lambda x: 'garage construction' in (x or []))]

# show the result
positive_garage_specialists[['Company Name', 'builder_names', 'cleaned_review', 'sentiment', 'sentiment_score']].head()
```

```python
# calculate the average sentiment analysis score each company
builder_scores = df.groupby('Company Name')['sentiment_score'].mean().reset_index()

# sort the score with ascending. Higher score of companies are above.
top_builders = builder_scores.sort_values(by='sentiment_score', ascending=False)

# check the order
top_builders.head(30)
```

```python
from transformers import pipeline, AutoTokenizer

# load the pipeline of sentiment analysis for Hugging Face
sentiment_pipeline = pipeline('sentiment-analysis')

# load tokenizer and restrict the token up to 512
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased", model_max_length=512, truncation=True)

# function for triming the text to 512 token
def truncate_text(text, max_tokens=512):
    tokens = tokenizer(text, truncation=True, max_length=max_tokens, return_tensors="pt")
    return tokenizer.decode(tokens["input_ids"][0], skip_special_tokens=True)

# implement sentiment analysis with triming texts
df['cleaned_review'] = df['cleaned_review'].apply(truncate_text)
df['huggingface_sentiment'] = df['cleaned_review'].apply(lambda x: sentiment_pipeline(x)[0]['label'])
df['huggingface_score'] = df['cleaned_review'].apply(lambda x: sentiment_pipeline(x)[0]['score'])

# show the result
df[['cleaned_review', 'huggingface_sentiment', 'huggingface_score']].head()
```

```python
# save the data frame as CSV
df.to_csv('/content/drive/MyDrive/NLP_exercise_12252023/Sharpest_minds_Khaled/RealEstate2_all/text_based_reviews/merge_file/huggingface_review_with_sentiment.csv', index=False)

# show the result of sentiment analysis
df[['cleaned_review', 'huggingface_sentiment', 'huggingface_score']].head()
```

```python
# decide the threshhold. Let's say more than 0.7 is positive review.
threshold = 0.7

# extract builers which have only positive sentiment scores
positive_reviews = df[df['huggingface_score'] >= threshold]

# filtering with skill(positive review with the companies specialized garage construction for example)
positive_garage_specialists = positive_reviews[positive_reviews['skills'].apply(lambda x: 'garage construction' in (x or []))]

# show the result
positive_garage_specialists[['Company Name', 'builder_names', 'cleaned_review', 'huggingface_sentiment', 'huggingface_score']].head()
```

```python
# calculate the average sentiment analysis score each company
builder_scores = df.groupby('Company Name')['huggingface_score'].mean().reset_index()

# sort the score with ascending. Higher score of companies are above.
top_builders = builder_scores.sort_values(by='huggingface_score', ascending=False)

# check the order
top_builders.head(30)
```
