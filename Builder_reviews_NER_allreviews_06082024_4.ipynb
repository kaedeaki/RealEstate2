```python
from flair.data import Sentence
from flair.nn import Classifier
from flair.models import SequenceTagger
 #Load the model
tagger = SequenceTagger.load('ner')
sentence = Sentence('George Washington went to Washington.')
tagger.predict(sentence)

print(sentence.to_tagged_string())

for entity in sentence.get_spans('ner'):
    print(entity)
```

```python
#function for NER
def apply_ner(text):
    sentence = Sentence(text)
    tagger.predict(sentence)
    entities = sentence.to_dict(tag_type='ner')['entities']
    return entities
```

```python
df['Review_Text'] = str(df['Review Text'])
```

```python
#apply to entire dataset
df['entities'] = df['Review Text'].apply(apply_ner)

#save the result
df.to_csv('/content/drive/MyDrive/NLP_exercise_12252023/Sharpest_minds_Khaled/RealEstate2_all/text_based_reviews/merge_file/ner_result.csv', index=False)
```

```python
def getsentimentanalysis(sentence, neutral_threshold=0.5):
    """
    Analyzes the sentiment of a given sentence.

    Args:
    - sentence (str): The input sentence for sentiment analysis.
    - neutral_threshold (float): The confidence threshold for classifying as "NEUTRAL".

    Returns:
    - sentiment (str): The predicted sentiment (positive, negative, neutral).
    - confidence (float): The confidence score associated with the sentiment prediction.
    """
    if not isinstance(sentence, str) or sentence.strip() == "":
        return "", ""

    # Create a Sentence object
    sample = flair.data.Sentence(str(sentence))

    # Predict sentiment
    sentiment_model.predict(sample)

    # Get sentiment and confidence
    top_label = sample.labels[0]
    sentiment = top_label.value
    confidence = top_label.score

    # If the top label is not "NEUTRAL" and its confidence is below the threshold, set sentiment as "NEUTRAL"
    if sentiment != "NEUTRAL" and confidence < neutral_threshold:
        sentiment = "NEUTRAL"

    return sentiment, confidence

#create the space to put sentiments and confidences
sentiments = []
confidences = []

for sentence in df['Review Text']:
    s, c = getsentimentanalysis(sentence)
    sentiments.append(s)
    confidences.append(c)
```
