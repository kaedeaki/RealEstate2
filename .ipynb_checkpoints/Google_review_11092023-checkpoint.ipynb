```python
import csv
import time
from selenium import webdriver
from parsel import Selector

chromedriver_path = './chromedriver'
driver = webdriver.Chrome(chromedriver_path)

csv_file_path = r'C:\Users\Madoka\Desktop\NLP_RealEstate\projects\ohba\builder_info_1.csv'
with open(csv_file_path, 'r', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    companies_info = [{'web': row['web']} for row in csv_reader]

results = []
for company_info in companies_info:
    web = company_info['web']
    search_query = f'{web} site:google.com/maps'
    url = f'https://www.google.com/search?q={search_query}'
    driver.get(url)
    time.sleep(5)

    page_content = driver.page_source
    response = Selector(page_content)

    reviews = []
    for el in response.xpath('//div[@class="qs41qe"]/div[@class="fm06If"]/div[contains(@class, "gm2-body-2")]/div'):
        reviews.append({
            'title': el.xpath('.//div[@class="m1e9Pa"]/span[@class="FcUp2b"]/text()').extract_first(''),
            'rating': el.xpath('.//span[contains(@aria-label, "Rated")]/@aria-label').extract_first('').strip(),
            'body': el.xpath('.//span[@jsinstance="0"]/span[@class="Yelper__review-text"]/text()').extract_first(''),
        })

    results.append({
        'web': web,
        'reviews': reviews,
    })

output_csv_file_path = './results.csv'
with open(output_csv_file_path, 'w', encoding='utf-8', newline='') as output_csv_file:
    fieldnames = ['web', 'reviews']
    csv_writer = csv.DictWriter(output_csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()
    csv_writer.writerows(results)

driver.quit()
```
